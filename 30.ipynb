```python
#clustring
import pandas as pd
from sklearn.cluster import KMeans
data ={
    "age":[22,25,24,45,50,48,20,47],
    "income":[22000,25000,24000,45000,50000,48000,20000,47000]
}

df=pd.DataFrame(data)
print(df)

#input / feature
x=df[['age','income']]

#model
model=KMeans(n_clusters=2,random_state=42)

model.fit(x)

#clusters
clusters=model.labels_ # 0=youg age/1=older age
print(clusters)
df['cluster']=clusters #added new column
print(df)
```

```python
#example
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
data ={
    "age":[22,25,24,45,50,48,20,47],
    "income":[22000,25000,24000,45000,50000,48000,20000,47000]
}

df=pd.DataFrame(data)
print(df)

x=df[['age','income']]
model=KMeans(n_clusters=2,random_state=42)
model.fit(x)

df['cluster']=model.labels_

plt.scatter(df['age'],df['income'],c=df['cluster']) #c= array
plt.xlabel('age')
plt.ylabel('income')
plt.title('Grouped data')
plt.show()



```

```python
#on example of clustring with matplotlib
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data={
    "height":[155,140,132,167,180],
    "weight":[50,60,70,80,90],

}
df=pd.DataFrame(data)
print(df)

x=df[['height','weight']]
model=KMeans(n_clusters=2,random_state=42)
model.fit(x)

df['cluster']=model.labels_
print(df)

plt.bar(df['height'],df['cluster'])
plt.xlabel('height')
plt.ylabel('weight')
plt.title('information')
plt.show()
#
```
